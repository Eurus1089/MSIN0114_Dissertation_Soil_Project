```python
# Import libraries used in this notebook
import numpy as np
import pandas as pd
```

```python
# Load the data frames
summary_nasa_bogura = pd.read_csv("Nasa_summary_bogura.csv")
summary_nasa_cumilla = pd.read_csv("Nasa_summary_cumilla.csv")
summary_nasa_muktagacha = pd.read_csv("Nasa_summary_muktagacha.csv")
```

```python
# View the data frames
summary_nasa_bogura.head(5)
summary_nasa_cumilla.head(5)
summary_nasa_muktagacha.head(5)
```

```python
# Check all 3 data frames have the same columns
all_columns_match = (
    summary_nasa_bogura.columns.equals(summary_nasa_cumilla.columns) and
    summary_nasa_cumilla.columns.equals(summary_nasa_muktagacha.columns)
)
print(f"All columns identical: {all_columns_match}")
```

```python
# Combine 3 data frames (vertical stacking)
summary_nasa_all = pd.concat(
    [summary_nasa_bogura, summary_nasa_cumilla, summary_nasa_muktagacha],
    axis=0,
    ignore_index=True  # Reset index
)
```

```python
# Verify the combination result
print(f"Total rows: {len(summary_nasa_all)}")
print(f"District counts:\n{summary_nasa_all['location'].value_counts()}")
```

```python
# Export the processed DataFrame to CSV
summary_nasa_all.to_csv('NASA_summary_all.csv', index=False)
```
